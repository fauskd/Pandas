```python
python_df
```

```python
python_df.rename(columns = {'Country':'country_respond', 'LanguageHaveWorkedWith':'knows_python'}, inplace = True)
```

```python
python_df
```

```python
python_df['pctknowspython'] = (python_df['knows_python']/python_df['country_respond'])*100
```

```python
python_df.sort_values(by = 'pctknowspython', ascending = False, inplace = True)
```
